```python
# Standard library
import warnings
from operator import truediv
from pathlib import Path

# Data manipulation
import numpy as np
import pandas as pd

ITEMS_DATA_PATH = "../../data/raw/items.csv"
TRAIN_DATA_PATH = "../../data/raw/train.csv"
```

```python
items = pd.read_csv(ITEMS_DATA_PATH, sep="|")
items.sample(10, random_state=42)
```

```python
items.info(show_counts=True)
```

```python
items.nunique()
```

```python
#checking if pid has duplicate
items_duplicate_pid = items["pid"].duplicated().sum()
print(f"items_duplicate_pid: {items_duplicate_pid}")

#checking for 0 and NA
items_missing_pid = items["pid"].isna().sum()
print(f"items_missing_pid: {items_missing_pid}")

#number of uniques
items_unique_pid = items["pid"].nunique()
print(f"items_unique_pid: {items_unique_pid}")
```

```python
#checking if pid has duplicate
items_size_manufacturer = items["manufacturer"].size
print(f"items_size_manufacturer: {items_size_manufacturer}")

#checking unique values
items_unique_manufacturer = items["manufacturer"].nunique()
print(f"items_unique_manufacturer: {items_unique_manufacturer}")

#check for missing values
items_missing_manufacturer = items["manufacturer"].isna().sum()
print(f"items_missing_manufacturer: {items_missing_manufacturer}")
```
